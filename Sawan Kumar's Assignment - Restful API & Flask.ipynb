# 23. What is the role of Flask’s session object ?
>* User Session Management:
>>* The session object stores data that should be preserved across requests for the same user. For example, user authentication info (like login status), preferences, or temporary data can be stored in the session.
>>* The session is persistent across different pages and requests within a session.

>* Client-Side Storage (Secure):
>>* Flask’s session uses cookies to store the session data on the client-side (in the user's browser).
>>* Data is stored as a cookie that is sent along with every request. This cookie is signed using a secret key to ensure that it cannot be tampered with by the client.

>* Security:
>>* The session is signed (using app.secret_key), which means its content can’t be modified by the client. If the cookie is modified or tampered with, it becomes invalid and Flask will reject it.
>>* You can use secure cookies to make sure the session data is encrypted if needed.

>* Temporary Data Storage:
>>* It is commonly used to store data temporary across requests for individual users, such as user login information, form data, flash messages, or other personalized settings.

```python
# 7. How can you validate form data in Flask ?

from flask import Flask, render_template, request, flash, redirect, url_for
from wtforms import Form, StringField, IntegerField, EmailField, validators

app = Flask(__name__)
app.secret_key = "secret"  # For flash messages.  **Important:  Use a strong, random key in production!**

# Define a form class.  This replaces WTForms' deprecated DataRequired with InputRequired.
class MyForm(Form):
    name = StringField("Name", [validators.InputRequired("Name is required!")])
    age = IntegerField("Age", [
        validators.InputRequired("Age is required!"),
        validators.NumberRange(min=0, max=150, message="Age must be between 0 and 150.")
    ])
    email = EmailField("Email", [
        validators.InputRequired("Email is required!"),
        validators.Email("Invalid email address!")
    ])
    # Example of a custom validation function
    def validate_name(self, field):
        if len(field.data) < 3:
            raise validators.ValidationError("Name must be at least 3 characters long.")

@app.route("/", methods=["GET", "POST"])
def index():
    form = MyForm(request.form)
    if request.method == "POST":
        if form.validate():
            # If the form is valid, process the data and redirect.
            name = form.name.data
            age = form.age.data
            email = form.email.data
            flash(f"Data submitted successfully! Name: {name}, Age: {age}, Email: {email}")
            return redirect(url_for("success"))  #  Redirect to a success page
        else:
            # If the form is invalid, the errors will be available in form.errors.
            # No need to explicitly pass them to the template; Flask-WTF does that.
            flash("Form has errors. Please correct them.") # Add flash
    return render_template("index.html", form=form) # Render the form, passing the form object.

@app.route("/success")
def success():
    return render_template("success.html")

if __name__ == "__main__":
    app.run(debug=True) # Set debug=False in production.
```

```python
# 8. How do you manage sessions in Flask ?

from flask import Flask, session, redirect, url_for
from datetime import timedelta

app = Flask(__name__)
# In a real application, you would use a more secure secret key
#  and store it in a configuration file or environment variable.
app.secret_key = "super secret key"
app.permanent_session_lifetime = timedelta(minutes=30)  # Optional: Set session lifetime

@app.route('/')
def index():
    """
    This function handles the root route of the web application.  It checks
    if a username is stored in the session.  If the username exists, it
    greets the user; otherwise, it redirects the user to the login page.
    """
    if 'username' in session:
        username = session['username']
        return f"Logged in as {username}<br><a href='/logout'>Logout</a>"
    return "You are not logged in <br><a href='/login'>Login</a>"

@app.route('/login', methods=['GET', 'POST'])
def login():
    """
    This function handles the login route, supporting both GET and POST requests.
    -   GET:  Displays the login form (a simple string in this example).
    -   POST: Processes the login form submission.  It sets the 'username'
        in the session and optionally sets the session to be permanent.
    """
    from flask import request  # Import request here to avoid potential circular import
    if request.method == 'POST':
        username = request.form['username']
        session['username'] = username
        if 'remember_me' in request.form:
            session.permanent = True  # Make the session permanent.
        return redirect(url_for('index'))
    return """
        <form method="post">
            <p><input type=text name=username>
            <p><input type=checkbox name=remember_me> Remember Me
            <p><input type=submit value=Login>
        </form>
    """

@app.route('/logout')
def logout():
    """
    This function handles the logout route.  It removes the 'username'
    from the session, effectively logging the user out.
    """
    session.pop('username', None)  # Use pop to avoid KeyError if 'username' doesn't exist
    return redirect(url_for('index'))

if __name__ == '__main__':
    #  Use a port number > 1024, if running without root privileges.
    app.run(debug=True, port=5001)
```

```python
# 12. How do you define a custom Jinja filter in Flask ?

from flask import Flask, render_template

app = Flask(__name__)

# Define a custom Jinja filter
@app.template_filter('reverse_string')
def reverse_string_filter(s):
    """Custom Jinja filter to reverse a string."""
    if isinstance(s, str):
        return s[::-1]  # Reverses the string
    return s

@app.route('/')
def home():
    text = "Hello, Flask!"
    return render_template('home.html', text=text)

if __name__ == '__main__':
    app.run(debug=True)
```

```python
# 13. How can you redirect with query parameters in Flask ?

from flask import Flask, redirect, url_for, request

app = Flask(__name__)

@app.route('/')
def home():
    return "Welcome to the home page! <a href='/redirect-example'>Go to redirect example</a>"

@app.route('/redirect-example')
def redirect_example():
    # Redirecting with query parameters
    return redirect(url_for('target_page', param1='value1', param2='value2'))

@app.route('/target')
def target_page():
    # Accessing query parameters
    param1 = request.args.get('param1')
    param2 = request.args.get('param2')
    return f"Redirected with param1: {param1} and param2: {param2}"

if __name__ == '__main__':
    app.run(debug=True)
```

```python
# 14. How do you return JSON responses in Flask ?

from flask import Flask, jsonify

app = Flask(__name__)

@app.route('/')
def home():
    return "Welcome to the JSON response example!"

@app.route('/json-example')
def json_example():
    # Create a dictionary to return as JSON
    response_data = {
        "name": "John Doe",
        "age": 30,
        "city": "New York"
    }

    # Return JSON response using jsonify()
    return jsonify(response_data)

if __name__ == '__main__':
    app.run(debug=True)
```

```python
# 15. How do you capture URL parameters in Flask ?

from flask import Flask, request

app = Flask(__name__)

# Route capturing a URL parameter
@app.route('/user/<username>')
def show_user(username):
    return f"Hello, {username}!"

# Route capturing multiple URL parameters
@app.route('/post/<int:post_id>/comment/<int:comment_id>')
def show_post_comment(post_id, comment_id):
    return f"Post ID: {post_id}, Comment ID: {comment_id}"

# Route capturing query parameters
@app.route('/search')
def search():
    # Capturing query parameters using request.args.get()
    query = request.args.get('query')
    page = request.args.get('page', default=1, type=int)  # Default to page 1 if not provided
    return f"Search for: {query}, Page: {page}"

if __name__ == '__main__':
    app.run(debug=True)
```
